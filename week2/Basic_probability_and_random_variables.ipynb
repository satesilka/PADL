# Gaussian distributions

Let's plot out some Gaussian *probability density functions*.

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
x = np.arange(-5,5,0.01)
plt.plot(x,norm.pdf(x,0,1),label=r'$\mu=0,\sigma^{2}=1$')
plt.plot(x,norm.pdf(x,2,3),label=r'$\mu=2,\sigma^{2}=1$')
plt.plot(x,norm.pdf(x,0,2),label=r'$\mu=0,\sigma^{2}=2$')
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.title("Some Gaussian distributions (pdf)")
plt.legend()
plt.show()
```

We can also plot the *cumulative density functions* which serves to show that
in each case the area under the pdf is 1.

```python
plt.plot(x,norm.cdf(x,0,1),label=r'$\mu=0,\sigma^{2}=1$')
plt.plot(x,norm.cdf(x,2,1),label=r'$\mu=2,\sigma^{2}=1$')
plt.plot(x,norm.cdf(x,0,2),label=r'$\mu=0,\sigma^{2}=2$')
plt.xlabel(r'$x$')
plt.ylabel(r'$F(x)$')
plt.title("Some Gaussian distributions (cdf)")
plt.legend()
plt.show()
```

```python
plt.plot(x,norm.pdf(x,0,1),label=r'$x$')
plt.plot(x,norm.pdf(x,2,1),label=r'$x+2$')
plt.plot(x,norm.pdf(x,0,4),label=r'$2x$')
plt.xlim(-3,8)
plt.ylim(0,0.5)
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.title("Some functions of a standard Gaussian distribution (pdf)")
plt.legend()
plt.show()
```

```python
from mpl_toolkits import mplot3d
from scipy.stats import multivariate_normal
x = np.linspace(-2, 2, 30)
y = np.linspace(-2, 2, 30)
X, Y = np.meshgrid(x, y)
rv = multivariate_normal([0.0, 0.0], [[2.0, -0.3], [-0.3, 0.5]])
pos = np.dstack((X, Y))
Z = rv.pdf(pos)
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap = 'binary')
ax.view_init(70,30)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('pdf')
```
