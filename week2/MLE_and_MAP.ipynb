# Maximum likelihood estimation

The log-likelihood for the $r=12$ heads out of $n=20$ coin tosses is $$\log L(\theta) = \log {20 \choose 12} + 12 \log \theta + 8 \log (1-\theta)$$ Let's plot it (but leave out the constant $ \log {20 \choose 12}$ term).

```python
import numpy as np
from math import log
import matplotlib.pyplot as plt
import matplotlib
#matplotlib.rcParams['text.usetex'] = True
x = np.arange(0.01,1,0.01) #can't start at 0 would get an error
def llh(theta):
    return 12*log(theta) + 8*log(1-theta)
y = np.array([llh(theta) for theta in x])
plt.plot(x,y)
plt.axvline(x=0.6,color='red')
plt.xlabel(r'$\theta$',fontsize=20)
plt.ylabel(r'$\log L(\theta)$',fontsize=20)
plt.show()
```

OK, so one can well believe that $\hat{\theta}= \frac{12}{20} = 0.6$ is the maximal value of this function, but we need to be sure! We have $$\frac{d\log L(\theta)}{d\theta} = \frac{12}{\theta} - \frac{8}{1-\theta}$$

$\hat{\theta}= \frac{12}{20} = 0.6$ is the only solution to  $\frac{12}{\theta} - \frac{8}{1-\theta} = 0$ so we have a *stationary point* at 0.6. The second derivative $$\frac{d^{2}\log L(\theta)}{d\theta^{2}} = -\frac{12}{\theta^{2}} - \frac{8}{(1-\theta)^{2}}$$ is negative for all values of $\theta \in (0,1)$ so the likelihood function is *concave* ( as you can see from the graph!) and so 0.6 is the global maximum.

We don't really need them but here are the graphs of the first and second derivatives of the log-likelihood:

```python
x = np.arange(0.01,1,0.01) #can't start at 0 would get an error
def dllh(theta):
    return 12/theta + 8/(1-theta)
y = np.array([dllh(theta) for theta in x])
plt.plot(x,y)
plt.xlabel(r'$\theta$',fontsize=20)
plt.ylabel(r'$\frac{d\log L(\theta)}{d\theta}$',fontsize=20)
plt.show()
```

```python
x = np.arange(0.01,1,0.01) #can't start at 0 would get an error
def d2llh(theta):
    return -12/theta**2 - 8/((1-theta)**2)
y = np.array([d2llh(theta) for theta in x])
plt.plot(x,y)
plt.xlabel(r'$\theta$',fontsize=20)
plt.ylabel(r'$\frac{d^{2}\log L(\theta)}{d\theta^{2}}$',fontsize=20)
plt.show()
```

# Beta distributions

Let's plot a few Beta distributions.

```python
from scipy.stats import beta
x = np.arange(0,1,0.01)
plt.plot(x,beta.pdf(x,2,3),label=r'$\alpha=2,\beta=3$')
plt.plot(x,beta.pdf(x,1,1),label=r'$\alpha=1,\beta=1$')
plt.plot(x,beta.pdf(x,5,3),label=r'$\alpha=5,\beta=3$')
plt.xlabel(r'$\theta$',fontsize=20)
plt.ylabel(r'$\mathrm{Beta}(\theta|\alpha,\beta)$',fontsize=20)
plt.legend()
plt.show()
```

If we have values of $\alpha$ and $\beta$ below 1 then we get 'U-shaped' distributions:

```python
plt.plot(x,beta.pdf(x,0.4,0.3),label=r'$\alpha=0.4,\beta=0.3$')
plt.xlabel(r'$\theta$',fontsize=20)
plt.ylabel(r'$\mathrm{Beta}(\theta|\alpha,\beta)$',fontsize=20)
plt.legend()
plt.show()
```
